```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

## Linear Model Plots

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'])
```

Or instead of calling the hue parameter one can use something simlar to subplots.

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
```

### Another example:

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', row='time', hue='sex', markers=['o','v'], aspect=0.6, height=10)
```
